```python
import plotly.express as px
from jupyter_dash import JupyterDash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import dash_bootstrap_components as dbc

df = pd.read_csv('https://raw.githubusercontent.com/ccodwg/Covid19Canada/master/timeseries_prov/active_timeseries_prov.csv')
df.head()
```

```python
dates = df['date_active'].unique()
province_names = df['province'].unique()
def provinceData(name):
    return df.loc[df['province'] == name]

fig = px.line(provinceData('Ontario'), x='date_active', y=['cumulative_cases', "cumulative_recovered", "cumulative_deaths"])
fig_AC = px.line(provinceData('Ontario'), x='date_active', y="active_cases")
fig_CC = px.bar(provinceData('Ontario'), x='date_active', y="cumulative_cases")
fig_CR = px.bar(provinceData('Ontario'), x='date_active', y="cumulative_recovered")
fig_CD = px.bar(provinceData('Ontario'), x='date_active', y="cumulative_deaths")

fig_AC.show()
fig_CC.show()
fig_CR.show()
fig_CD.show()
```
